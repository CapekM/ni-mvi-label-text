```python
import json
```

```python
data = []
```

```python
for file in ['data_100','data_500', 'data_1000', 'data_2371', 'data_3000','data_5000']:
    f = open('../crawler/results/'+file+'.json', encoding="utf8") 

    data1 = json.load(f) 
    print(len(data1))
    for i in data1:
        flag = True
        for x in data:
            if(i['website'] == x['website']):
                flag = False
                break
        if(flag):
            data.append(i)


    # Closing file 
    f.close()
len(data)
```

```python
labels = {}
for dic in data:
    for k in dic['labels']:
        labels[k] = labels.get(k, 0) + 1
```

```python
# Similiar labels

similiar_labels = {}
for l in labels:
    for l2 in labels:
        if l in l2 and (l != l2):
            similiar_labels[l2] = l
            print(l + "  " + l2)
```

```python
# Unite labels
cnt = 0
for d in similiar_labels.keys():
    for dic in data:
        flag = False
        for label in dic['labels']:
            if d in label:
                flag = True
        if flag and similiar_labels[d] not in dic['labels']:
            cnt +=1
            dic['labels'].append(similiar_labels[d])
cnt
```

```python
labels = {}
for dic in data:
    for k in dic['labels']:
        labels[k] = labels.get(k, 0) + 1
```

```python
# Unite Coronavirus labels - there is Coronavirus pandemic, Coronavirus lockdown measures, Coronavirus testing
for word in ['Coronavirus', 'Police']:
    for dic in data:
        flag = False
        for label in dic['labels']:
            if word in label:
                flag = True
        if flag:
            dic['labels'] = [i for i in dic['labels'] if word not in i]
            dic['labels'].append(word)
```

```python
print(len(labels))
# delete insignificant labels
to_del = [key for key in labels if labels[key] < 100]

for key in to_del: del labels[key]

len(labels)
```

```python
# delete labels
data2 = data
for dic in data2:
    dic['labels'] = [i for i in dic['labels'] if i not in to_del]

# delete articles without label
data2 = [item for item in data2 if item['labels']]
len(data2)
```

```python
# How many articles with just one label
cnt = 0
for dic in data2:
    if len(dic['labels']) == 1:
        cnt +=1
cnt
```

```python
labels = {}
for dic in data2:
    for k in dic['labels']:
        labels[k] = labels.get(k, 0) + 1
len(labels)
```

```python
sorted = {key: value for key, value in sorted(labels.items(), key=lambda item: item[1])}
sorted
```
